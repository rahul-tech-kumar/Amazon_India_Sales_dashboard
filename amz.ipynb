```python
import mysql.connector
import pandas as pd

# Load your CSV
df = pd.read_csv("cleaned_data.csv")

# Connect to MySQL
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="12345",
    database="amazon_sales"
)
print("Connected to My sql")
cursor = conn.cursor()
```

```python
create_table_query = """
CREATE TABLE IF NOT EXISTS transactions (
    transaction_id VARCHAR(50) PRIMARY KEY,
    customer_id VARCHAR(50),
    product_id VARCHAR(50),
    product_name VARCHAR(255),
    category VARCHAR(100),
    subcategory VARCHAR(100),
    brand VARCHAR(100),
    original_price_inr FLOAT,
    discount_percent FLOAT,
    discounted_price_inr FLOAT,
    quantity INT,
    subtotal_inr FLOAT,
    delivery_charges FLOAT,
    final_amount_inr FLOAT,
    customer_city VARCHAR(100),
    customer_state VARCHAR(100),
    customer_tier VARCHAR(50),
    customer_spending_tier VARCHAR(50),
    customer_age_group VARCHAR(50),
    payment_method VARCHAR(50),
    delivery_days INT,
    delivery_type VARCHAR(50),
    is_prime_member BOOLEAN,
    is_festival_sale BOOLEAN,
    festival_name VARCHAR(100),
    customer_rating FLOAT,
    return_status VARCHAR(50),
    order_month INT,
    order_year INT,
    order_quarter INT,
    product_weight_kg FLOAT,
    is_prime_eligible BOOLEAN,
    product_rating FLOAT,
    corrected_price_inr FLOAT
);
"""
cursor.execute(create_table_query)
print("Table created or already exists.")
```

```python
insert_query = """
INSERT INTO transactions (
    transaction_id, customer_id, product_id, product_name, category, subcategory, brand,
    original_price_inr, discount_percent, discounted_price_inr, quantity, subtotal_inr, delivery_charges,
    final_amount_inr, customer_city, customer_state, customer_tier, customer_spending_tier,
    customer_age_group, payment_method, delivery_days, delivery_type, is_prime_member,
    is_festival_sale, festival_name, customer_rating, return_status, order_month, order_year,
    order_quarter, product_weight_kg, is_prime_eligible, product_rating, corrected_price_inr
) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""
```

```python
# Drop order_date column
df = df.drop(columns=["order_date"])

# Replace NaN with None for MySQL
df = df.where(pd.notnull(df), None)

# Convert booleans to Python bool
bool_cols = ['is_prime_member', 'is_festival_sale', 'is_prime_eligible']
for col in bool_cols:
    df[col] = df[col].astype(bool)

# Convert to list of tuples
data = [tuple(row) for row in df.to_numpy()]
```

```python


# Insert all rows in one batch (faster for large datasets)
batch_size = 10000  # insert 10k rows at a time to avoid memory issues
for i in range(0, len(data), batch_size):
    cursor.executemany(insert_query, data[i:i+batch_size])
    conn.commit()
    print(f"Inserted rows {i} to {i+batch_size}")

print("All data inserted successfully!")
conn.close()
```
